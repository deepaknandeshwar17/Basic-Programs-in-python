```python
def count_digit(n):
    num = n
    count = 0
    step = 0

    while num > 0:
        count += 1
        step += 1
        # print(f"count at step {step} is {count}")
        print(f"num at the step {step} is {num}")
        num = num // 10
        

    return count

print(count_digit(54321))
```

```python
def reverse_number(n):
    num = n
    rev = 0

    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num = num // 10
    return rev

print(reverse_number(12345))
```
